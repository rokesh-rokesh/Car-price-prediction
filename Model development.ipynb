# Model development

```python
import pandas as pd
import numpy as np
import seaborn as sns
df = pd.read_pickle("cleaned_car_data.csv")
```

```python
class Linear_Regression:
    
    def model(self,X,y):
        X_mean,y_mean = np.mean(X),np.mean(y)
        X_mean_diff, y_mean_diff = X-X_mean, y-y_mean
        self.b1 = (X_mean_diff @ y_mean_diff) / (X_mean_diff @ X_mean_diff)
        self.b0 = y_mean - (self.b1 * X_mean)
        return self
    
    def predict(self,X):
        return self.b1 * X + self.b0
```

```python
if __name__ == "__main__":
    Engine_size,Price = df["engine-size"],df["price"]
    X = np.array(Engine_size,dtype = int)
    y = np.array(Price,dtype = int)
    lr = Linear_Regression()
    lr.model(X,y)
    #train the model by engine-size
    y_hat = lr.predict(X)
    #test the model
    y_hat_1 = lr.predict(282)
    y_hat_2 = lr.predict(100)
    print("Price of car when engine-size is 282:",y_hat_1)
    print("Price of car when engine-size is 100:",y_hat_2)
```

So,we conclude the price of car by engine-size because it has strong positive linear relation
